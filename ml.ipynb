```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("AIDS_Classification.csv")
```

```python
df.head()
```

### Data Cleaning

```python
for i in df.keys():
  print(i,df[i].nunique())
```

```python
df.corr()
```

```python
df.drop(["time","trt","hemo","drugs","karnof","race","treat","cd40","cd420","oprior","homo","str2","symptom"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.drop('infected',axis=1)
y = df['infected']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
classifer = GaussianNB()
# Train model
model = classifer.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
sample = x_test.iloc[0]
dic = {0:"AIDS not found",1:"AIDS found"}
print(dic[pred[0]])
```

```python
from sklearn.metrics import accuracy_score, classification_report
print(classification_report(y_test,pred))
accuracy_score(y_test,pred)
```
